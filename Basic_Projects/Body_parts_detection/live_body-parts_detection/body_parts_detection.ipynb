```python
DIRECTORY = r'/Users/sureshkalisetty/Downloads/AI&ML/mine prac/pycharm/real_projects/Body_parts_detection/Body_parts_detection'
CATEGORIES = os.listdir('/Users/sureshkalisetty/Downloads/AI&ML/mine prac/pycharm/real_projects/Body_parts_detection/Body_parts_detection')
CATEGORIES = [cat for cat in CATEGORIES if cat != '.DS_Store']

data = []
for category in CATEGORIES:
    folder = os.path.join(DIRECTORY, category)

    label = CATEGORIES.index(category)
    
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        img_array = cv2.imread(img_path)
        img_array = cv2.resize(img_array, (255,255))

        data.append([img_array, label])
```

```python
x, y = zip(*data)

x = np.array(x)
x = x/255
y = np.array(y)

print(f'x : {x.shape}')
print(f'y : {y.shape}')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=45)

print(f'x_train : {x_train.shape}')
print(f'y_train : {y_train.shape}')
print(f'x_test : {x_test.shape}')
print(f'y_test : {y_test.shape}')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.3,
    zoom_range=0.3,
    width_shift_range=0.3,
    height_shift_range=0.3,
    horizontal_flip=True,
    fill_mode='nearest')

data_generator = datagen.flow(x_train, y_train, batch_size=45)


img_aug, lab_aug = [],[]
for i in range(len(x_train)):
    image_a = x_train[i]
    image_a = image_a.reshape(1,255,255,3)
    label_a = y_train[i]

    data_generator1 = datagen.flow(x_train, batch_size=1)
    for j in range(2):
        image = next(data_generator1)[0]

        img_aug.append(image)
        lab_aug.append(label_a)

img_aug = np.array(img_aug)
lab_aug = np.array(lab_aug)

Images = np.concatenate([img_aug, x_train])
Labels = np.concatenate([lab_aug, y_train])

print(f'---------------------------------------')
print(f'x_train : {x_train.shape}')
print(f'y_train : {y_train.shape}')
print(f'---------------------------------------')
print(f'aug_img : {img_aug.shape}')
print(f'aug_lab : {lab_aug.shape}')
print(f'---------------------------------------')
print(f'Images : {Images.shape}')
print(f'Labels : {Labels.shape}')
print(f'---------------------------------------')       
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense

model1 = Sequential([
    Conv2D(64, (3,3), activation='relu', padding='same', input_shape=(255,255,3)),
    Conv2D(64, (3,3), activation='relu', padding='same'),
    MaxPooling2D((2,2), (2,2)),

    Conv2D(128, (3,3), activation='relu', padding='same', input_shape=(255,255,3)),
    Conv2D(128, (3,3), activation='relu', padding='same'),
    MaxPooling2D((2,2), (2,2)),

    Conv2D(64, (3,3), activation='relu', padding='same', input_shape=(255,255,3)),
    Conv2D(64, (3,3), activation='relu', padding='same'),
    MaxPooling2D((2,2), (2,2)),

    Flatten(),

    Dense(64, activation='relu'),
    Dense(10, activation='softmax')])


model1.compile(loss='sparse_categorical_crossentropy', 
               optimizer=tensorflow.keras.optimizers.legacy.Adam(learning_rate=0.001), 
               metrics=['accuracy'])
```

```python
early_stop = tensorflow.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0.0001,
    mode='auto',
    verbose=1,
    patience=20,
    baseline=None)

result1 = model1.fit(x_train, y_train, epochs=10, validation_data=(x_test,y_test), batch_size=45, callbacks=[early_stop])
```
